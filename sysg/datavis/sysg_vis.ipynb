```python
# Main FreeE 
fig = plt.figure(figsize=(8,4))
gs_left = plt.GridSpec(1, 3, wspace=0.20, top=1.00, left=0.05)
gs_right= plt.GridSpec(2, 3, hspace=0.30, wspace=0.20, top=1.00, left=0.15)
fax = fig.add_subplot(gs_left[0,:2])

def plot_free(data, color, trans=(None,None), deltas=(-0.05,0.05), s=0.001, excess=False):
    ls_meta = (0, (1.2,1.2))
    if trans[0]: start = trans[0]
    else:        start = 0.9
    if trans[1]: end   = trans[1]
    else:        end   = 2.1
    
    if not excess: 
        fit = UnivariateSpline(data[0], data[1], s=s)
        stable=np.arange(start, end+0.01, 0.01)
        plt.plot(stable, fit(stable), c=color, zorder=3)
        if trans[0]: 
            meta_L = np.arange(trans[0]+deltas[0], trans[0]+0.01, 0.01) 
            plt.plot(meta_L, fit(meta_L), c=color, ls=ls_meta, zorder=4)
            plt.plot([trans[0]]*2, [0.3, 3.5], ls=(0, (3,5)), c='k', lw=1.5)
        if trans[1]:
            meta_R = np.arange(trans[1], trans[1]+deltas[1]+0.01, 0.01)
            plt.plot(meta_R, fit(meta_R), c=color, ls=ls_meta, zorder=4)
            plt.plot([trans[1]]*2, [0.3, 3.5], ls=(0, (3,5)), c='k', lw=1.5)
    if excess: 
        fit, ref = UnivariateSpline(data[0]['Chi'], data[0]['F'], s=s), UnivariateSpline(data[1]['Chi'], data[1]['F'], s=s)
        stable=np.arange(start, end+1e-06, 1e-06)
        plt.plot(stable, fit(stable)-ref(stable), c=color, zorder=3)
        if trans[0]: 
            meta_L = np.arange(trans[0]+deltas[0], trans[0]+0.001, 0.001) 
            plt.plot(meta_L, fit(meta_L)-ref(meta_L), c=color, ls=ls_meta, zorder=4)
            plt.plot([trans[0]]*2, [0.3, 3.5], ls=(0, (3,5)), c='k', lw=1.5)
        if trans[1]:
            meta_R = np.arange(trans[1], trans[1]+deltas[1]+0.001, 0.001)
            plt.plot(meta_R, fit(meta_R)-ref(meta_R), c=color, ls=ls_meta, zorder=4)
            plt.plot([trans[1]]*2, [0.3, 3.5], ls=(0, (3,5)), c='k', lw=1.5)
    return

A = sysg.F4s02
t14,t43, t32, t21 = 1.108, 1.353, 1.5879, 1.9179
trans_kw = {'ls':(0, (3,5)), 'c':'k', 'lw':2.0}
plt.sca(fax)
plot_free((A['Chi']['p1'][2:8], A['F']['p1'][2:8]), cdict['p0c'], (None, t14))
plot_free((A['Chi']['p4'][2:12], A['F']['p4'][2:12]), cdict['p4c'], (t14, t43))
plot_free((A['Chi']['p3'][2:6], A['F']['p3'][2:6]), cdict['p3c'], (t43, t32))
plot_free((A['Chi']['p2'][4:], A['F']['p2'][4:]), cdict['p2c'], (t32, t21))
plot_free((A['Chi']['p1'][-4:], A['F']['p1'][-4:]), cdict['p1c'], (t21, None))

lab_kw = {'fontsize':18, 'ha':'center'}
shift = 1.55
plt.text(1.01, 0.30+shift, '(f)', **lab_kw)
plt.text(1.23, 0.85+shift, '(e)/(d)', **lab_kw)
plt.text(1.47, 1.18+shift, '(c)', **lab_kw)
plt.text(1.74, 1.40+shift, '(b)',  **lab_kw)
plt.text(2.01, 1.50+shift, '(a)', **lab_kw)

plt.xlim(0.8, 2.20)
plt.ylim(-1.5,3.5)
plt.xticks([1.0, 1.4, 1.8, 2.2])
plt.ylabel('$F$', fontsize=fs_axis)
plt.xlabel('$\chi$', fontsize=fs_axis)
tjy.ticks()

# Insets
ins1 = inset_axes(fax, width="100%", height="100%", loc="upper right",
                   bbox_to_anchor=(.15, .10, .20, .3), bbox_transform=fax.transAxes)
plt.sca(ins1) # left inset
Fref = A.iloc[0,:][:10]
plot_free((A.iloc[0,:], Fref), cdict['p0c'], (None,t14), excess=True, s=1e-07, deltas=(None, 0.005))
plot_free((A.iloc[3,:][:10], Fref), cdict['p4c'], (t14, None), excess=True, s=4e-07, deltas=(-0.003, None))
plt.plot([t14]*2, [-0.05, 0.05], **trans_kw)
ins1.set_ylabel(r'$F - F_{\mathrm{Swoll}}$')
ins1.set_xlim((1.100, 1.118))
ins1.set_ylim((-2e-04, 2e-04))
ins1.yaxis.set_ticks_position('right')
ins1.set_yticks([0])
ins1.set_xticks([])
tjy.ticks()

ins2 = inset_axes(fax, width="100%", height="100%", loc="upper right",
                   bbox_to_anchor=(.365, .325, .528, .35), bbox_transform=fax.transAxes)
plt.sca(ins2) # rght inset
Fref = A.iloc[1,:]
plot_free((A.iloc[3,], Fref), cdict['p4c'], (None, t43), excess=True)
plot_free((A.iloc[2,:], Fref), cdict['p3c'], (t43, t32), excess=True, deltas=(-0.05, 0.06))
plot_free((A.iloc[1,:], Fref), cdict['p2c'], (t32, t21), excess=True, deltas=(-0.06, 0.07))
plot_free((A.iloc[0,:], Fref), cdict['p1c'], (t21, None), excess=True, deltas=(-0.04, None))
plt.plot([t43]*2, [-0.05, 0.05], **trans_kw)
plt.plot([t32]*2, [-0.05, 0.05], **trans_kw)
plt.plot([t21]*2, [-0.05, 0.05], **trans_kw)

ins2.yaxis.set_label_position('right')
ins2.set_ylabel(r'$F - F_{\mathrm{2\ Cond}}$')
ins2.set_xlim((1.28, 2.02))
ins2.set_ylim((-0.022, 0.006))
ins2.set_yticks([0])
ins2.set_xticks([])

tjy.ticks()

# plt.savefig('imgs/free.pdf')
```

```python
rho_e_w = 3.35e23 # electron density of water, e/cm3 (Source: Brasla, 1985)
rho_e_w *= 1e2**3 / 1e10**3 # e/AA3
rho_e_p = 5.84e23 * 1e2**3 / 1e10**3 #Nafion = N_A * 2.00 g/cm3 * 264/544
rho_e_s = 6.99e23 * 1e2**3 / 1e10**3 #SiO2 = N_A * 2.32 g/cm3 * 30/60
r_e = 2.818e-15 # m
r_e *= 1e10 # AA

def RQZ_pha(phname, qz, GIT=GIT):
    def RQZ(z, dpdz, qzi): return np.power(np.absolute(np.trapz(dpdz*np.exp(1j*qzi*z), x=z)/rho_e_w), 2) # j is imaginary
    phname = GIT + phname
    ph = pd.read_csv(phname, sep="\s+", skiprows=0, header=None)
    pha = ph[2]
    # pha.update(pd.Series([2.32], index=[0])) # electron density of SiO2 substrate, assuming polymer electron density 1.0: (CH2)x = 0.95 (Source: Russel, 1990)
    pha.update(pd.Series([rho_e_s], index=[0])) # electron density of SiO2 substrate, assuming polymer electron density 1.0: (CH2)x = 0.95 (Source: Russel, 1990)
    phb = ph[10]
    phe = pha*rho_e_p + phb*rho_e_w # electron density profile, rho_e_protein = 1.0
    z = ph[0]*10 # nm to AA
    dz = z[1] - z[0]
    dpdz = np.gradient(phe, dz)
    return [RQZ(z, dpdz, i) for i in qz]

def RQZ_plot3(row, qz, filename, xc, light_frac, ls='-'):
    global step, delta
    if row == -1: 
        laxs.plot(qz, step0 + step*delta + np.array(RQZ_pha(filename, qz)), color=tjy.lighten_color(cdict['p'+xc+'c'], light_frac), alpha=1.0, ls=ls)
        step += 1
    else: laxs[row].plot(qz, RQZ_pha(filename, qz), color=tjy.lighten_color(cdict['p'+xc+'c'], light_frac), alpha=1.0)
    return
```
